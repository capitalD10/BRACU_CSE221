```python
inversion = 0
def merge(a_L, b_R):
  global inversion
  i = j = 0
  res = []
  while i < len(a_L) and j < len(b_R):
    if a_L[i] < b_R[j]:
        res.append(a_L[i])
        i += 1
    else:
        res.append(b_R[j])
        inversion += len(a_L) - i
        j += 1
  if i < len(a_L):
    res.extend(a_L[i:])
  elif j < len(b_R):
    res.extend(b_R[j:])
  return res

def mergeSort(arr):
  if len(arr) <= 1:
    return arr
  else:
    mid = len(arr) // 2
    left_arr = mergeSort(arr[:mid:])
    right_arr = mergeSort(arr[mid::])
    return merge(left_arr, right_arr)

N = int(input())
arr = list(map(int, input().split()))
merged = mergeSort(arr)
print(inversion)
print(*merged)
```
