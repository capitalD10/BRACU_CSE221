```python
def fasterFastMOD(a, b, mod):
  b_og = b
  a_og = a % mod
  counter = 1
  a = a_og
  if a == 1:
      return b % mod
  ans = a
  while b > 1:
      counter += counter
      ans = (ans + (ans * a)) % mod
      a = (a * a) % mod
      b //= 2
  if b_og > counter:
      ans = (ans + (a * fasterFastMOD(a_og, b_og - counter, mod))) % mod
  return ans

T = int(input())
cases = [list(map(int, input().split())) for x in range(T)]
res = []
for a, b, mod in cases:
  res.append(fasterFastMOD(a, b, mod))
print(*res, sep="\n")
```
