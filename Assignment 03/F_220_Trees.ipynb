```python
def tree220(inn, pree):
  indexes = {y: x for x, y in enumerate(inn)}
  preeIndex = [0]

  def helper(start, end):
    if start > end:
        return []
    value = pree[preeIndex[0]]
    preeIndex[0] += 1
    root = indexes[value]
    left_subtree = helper(start, root - 1)
    right_subtree = helper(root + 1, end)
    return left_subtree + right_subtree + [value]

  return helper(0, len(inn) - 1)
N = int(input())
a = list(map(int, input().split()))
b = list(map(int, input().split()))
print(*tree220(a, b))
```
