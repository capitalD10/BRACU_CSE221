```python
def binaryOrdering(arr):
  res = []
  def helper(low, high):
    if low > high:
        return
    mid = (low + high) // 2
    res.append(arr[mid])
    helper(low, mid - 1)
    helper(mid + 1, high)
  helper(0, len(arr) - 1)
  return res

N = int(input())
arr = list(map(int, input().split()))
print(*binaryOrdering(arr))
```
