```python
from collections import deque
N, M = map(int, input().split())
color = {x:0 for x in range(1, N + 1)}
edges = {x:[] for x in range(1, N + 1)}
result = []
for x in range(1, M + 1):
  key, value = map(int, input().split())
  edges[key].append(value)
  edges[value].append(key)
for node in edges:    # <-- This segment is optional
    edges[node].sort()
Q = deque()
color[1] = 1
Q.append(1)
while Q:
  vertex = Q.popleft()
  result.append(vertex)
  for x in edges[vertex]:
    if color[x] == 0:
      color[x] = 1
      Q.append(x)
for x in result:
  print(x, end=" ")
```
