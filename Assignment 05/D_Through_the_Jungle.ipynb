```python
from collections import deque
def ShortestBFS(S, D, edges, N):
  color = {x:0 for x in range(1, N + 1)}
  parent = {x:0 for x in range(1, N + 1)}
  Q = deque()
  color[S] = 1
  Q.append(S)
  while Q:
    vertex = Q.popleft()
    if vertex == D:
      break
    for elem in edges[vertex]:
      if color[elem] == 0:
        color[elem] = 1
        parent[elem] = vertex
        Q.append(elem)
  result = []
  last = D
  if color[D] == 0:
    return []
  while last != 0:
    result.append(last)
    last = parent[last]
  return result[::-1]

N, M, S, D, K = map(int, input().split())
edges = {x:[] for x in range(1, N + 1)}
for x in range(1, M + 1):
  node1, node2 = map(int, input().split())
  edges[node1].append(node2)
for node in edges:
  edges[node].sort()
Path1 = ShortestBFS(S, K, edges, N)
Path2 = ShortestBFS(K, D, edges, N)
if not Path1 or not Path2:
  print(-1)
else:
  total = Path1 + Path2[1:]
  print(len(total) - 1)
  print(*total)
```
