```python
from collections import deque
N, M, S, D = map(int, input().split())
node1 = list(map(int, input().split()))
node2 = list(map(int, input().split()))
edges = {x:[] for x in range(1, N + 1)}
color = {x:0 for x in range(1, N + 1)}
parent = {x:0 for x in range(1, N + 1)}
for x in range(M):
  edges[node1[x]].append(node2[x])
  edges[node2[x]].append(node1[x])
for node in edges:    # <-- This segment is optional
  edges[node].sort()
Q = deque()
color[S] = 1
Q.append(S)
while Q:
  vertex = Q.popleft()
  if vertex == D:
    break
  for elem in edges[vertex]:
    if color[elem] == 0:
      color[elem] = 1
      parent[elem] = vertex
      Q.append(elem)
if color[D] == 0:
  print(-1)
else:
  result = []
  last = D
  while last != 0:
    result.append(last)
    last = parent[last]
  print(len(result) - 1)
  for x in result[::-1]:
    print(x, end=" ")
```
