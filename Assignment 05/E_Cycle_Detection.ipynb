```python
import sys
sys.setrecursionlimit(2 * 100000 + 5)
def DFS(N, edges):
  color = {x:0 for x in range(1, N + 1)}
  def DFS_visit(vertex):
    color[vertex] = 1
    for x in edges[vertex]:
      if color[x] == 0:
        if DFS_visit(x):
          return True
      elif color[x] == 1:
        return True
    color[vertex] = 2
    return False
  for x in range(1, N + 1):
    if color[x] == 0:
      if DFS_visit(x):
        return True
  return False

N, M = map(int, input().split())
edges = {x:[] for x in range(1, N + 1)}
for x in range(1, M + 1):
  node1, node2 = map(int, input().split())
  edges[node1].append(node2)
for node in edges:
  edges[node].sort()
cycle = DFS(N, edges)
if cycle:
  print("YES")
else:
  print("NO")
```
