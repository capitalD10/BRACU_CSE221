```python
from collections import deque
def BFS(x, y, R, H):
  Q = deque()
  Q.append((x, y))
  color[x][y] = 1
  Dmnd = 0
  position = [(-1, 0), (1, 0), (0, -1), (0, 1)]
  while Q:
    x1, y1 = Q.popleft()
    if graph[x1][y1] == "D":
      Dmnd += 1
    for pos1, pos2 in position:
      move1, move2 = x1 + pos1, y1 + pos2
      if 0 <= move1 < R and 0 <= move2 < H:
        if color[move1][move2] == 0 and graph[move1][move2] != "#":
          color[move1][move2] = 1
          Q.append((move1, move2))
  return Dmnd

R, H = map(int, input().split())
graph = [list(input().strip()) for _ in range(R)]
color = [[0] * H for x in range(R)]
Diamonds = 0
for x in range(R):
  for y in range(H):
    if color[x][y] == 0 and graph[x][y] != "#":
      Diamonds = max(Diamonds, BFS(x, y, R, H))
print(Diamonds)
```
