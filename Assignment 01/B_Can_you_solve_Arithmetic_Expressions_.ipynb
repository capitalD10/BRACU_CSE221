```python
T = int(input())
operation = []
for x in range(T):
    N = input()
    N = N.split()
    if N[2] == "+":
        operation.append(int(N[1]) + int(N[3]))
    elif N[2] == "-":
        operation.append(int(N[1]) - int(N[3]))
    elif N[2] == "*":
        operation.append(int(N[1]) * int(N[3]))
    elif N[2] == "/":
        operation.append(int(N[1]) / int(N[3]))
for x in operation:
    print(f"{x:.6f}")
```
