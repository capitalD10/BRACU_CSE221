```python
N = int(input())
ID = input().split()
mark = input().split()
counter = 0
if len(ID) == N and len(mark) == N:
  std = [(int(ID[x]), int(mark[x])) for x in range(N)]
  for x in range(N):
      max_index = x
      for y in range(x + 1, N):
        if std[y][1] > std[max_index][1]:
            max_index = y
        elif std[y][1] == std[max_index][1] and std[max_index][0] > std[y][0]:
            max_index = y
      if max_index != x:
        std[x], std[max_index] = std[max_index], std[x]
        counter += 1
  print(f"Minimum swaps: {counter}")
  for x in std:
      print(f"ID: {x[0]} Mark: {x[1]}")
```
