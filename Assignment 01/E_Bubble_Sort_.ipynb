```python
def bubbleSort(arr, N):
    for i in range(N - 1):
        flag = False
        for j in range(N - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                flag = True
        if not flag:
            break
    print(*arr)
N = int(input().strip())
arr = input().strip().split()
if N != len(arr):
  print("Invalid")
else:
  res = [int(arr[x]) for x in range(N)]
  bubbleSort(res, N)
```
