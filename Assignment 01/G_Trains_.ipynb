```python
N = int(input())
trains = []
for x in range(N):
  inp = input().split()
  trains.append([inp[0], inp[len(inp) - 1], x, " ".join(inp)])
for x in range(N - 1):
  for y in range(N - x - 1):
    if trains[y][0] > trains[y + 1][0]:
      trains[y], trains[y + 1] = trains[y + 1], trains[y]
    elif trains[y][0] == trains[y + 1][0]:
      t1 = trains[y][1].split(":")
      t2 = trains[y + 1][1].split(":")
      if int(t1[0]) < int(t2[0]):
        trains[y], trains[y + 1] = trains[y + 1], trains[y]
      elif int(t1[0]) == int(t2[0]):
        if int(t1[1]) < int(t2[1]):
          trains[y], trains[y + 1] = trains[y + 1], trains[y]
for train in trains:
  print(train[3])
```
