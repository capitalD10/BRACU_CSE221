```python
import sys
sys.setrecursionlimit(3 * 100000 + 5)
def Kruskal(N, M, edges):
  edges.sort(key=lambda x: x[2])
  parent = [None] * (N + 1)
  size = [None] * (N + 1)
  for x in range(1, N + 1):
      parent[x] = x
      size[x] = 1
  cost = 0
  counter = 0
  def Find(X):
    if parent[X] != X:
      parent[X] = Find(parent[X])
    return parent[X]
  def Union(x, y, size):
    rootX = Find(x)
    rootY = Find(y)
    if rootX == rootY:
      return
    if size[rootX] < size[rootY]:
      rootX, rootY = rootY, rootX
    parent[rootY] = rootX
    size[rootX] += size[rootY]

  for node1, node2, weight in edges:
    if Find(node1) != Find(node2):
      Union(node1, node2, size)
      cost += weight
      counter += 1
      if counter == N - 1:
        break
  return cost

N, M = map(int, input().split())
edges = []
for x in range(M):
  edges.append(tuple(map(int, input().split())))
print(Kruskal(N, M, edges))
```
