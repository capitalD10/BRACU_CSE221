```python
N, M = list(map(int, input().split()))
node1 = list(map(int, input().split()))
node2 = list(map(int, input().split()))
indeg = [0] * N
outdeg = [0] * N
for x in range(M):
  outdeg[node1[x] - 1] += 1
  indeg[node2[x] - 1] += 1
result = [indeg[x] - outdeg[x] for x in range(N)]
for x in result:
  print(x, end=" ")
```
