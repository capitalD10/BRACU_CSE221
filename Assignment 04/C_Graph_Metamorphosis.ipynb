```python
N = int(input())
graph = [[0] * N for x in range(N)]
points = [list(map(int, input().split())) for x in range(N)]
for x in range(len(points)):
  for y in range(points[x][0]):
    graph[x][points[x][y + 1]] = 1
for x in range(N):
  for y in range(N):
    print(graph[x][y], end=" ")
  print()
```
