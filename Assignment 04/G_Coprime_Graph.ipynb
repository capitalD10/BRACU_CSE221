```python
import math
N, Q = list(map(int, input().split()))
query = [tuple(map(int, input().split())) for x in range(Q)]
graph = [[] for x in range(N + 1)]
for x in range(1, N + 1):
  for y in range(x + 1, N + 1):
    if (x != y) and (math.gcd(x, y) == 1):
      graph[x].append(y)
      graph[y].append(x)
for X, K in query:
  if K <= len(graph[X]):
      print(graph[X][K - 1])
  else:
      print(-1)
```
