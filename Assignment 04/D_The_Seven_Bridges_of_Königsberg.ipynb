```python
N, M = list(map(int, input().split()))
node1 = list(map(int, input().split()))
node2 = list(map(int, input().split()))
graph = [[] for x in range(N)]
for x in range(M):
  graph[node1[x] - 1].append([node2[x] - 1])
  graph[node2[x] - 1].append([node1[x] - 1])
eulerian = 0
for x in range(N):
    if len(graph[x]) % 2 != 0:
        eulerian += 1
if (eulerian == 2) or (eulerian == 0):
  print("YES")
else:
  print("NO")
```
