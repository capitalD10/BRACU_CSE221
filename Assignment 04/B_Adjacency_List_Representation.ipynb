```python
N, M = list(map(int, input().split()))
nod1 = list(map(int, input().split()))
nod2 = list(map(int, input().split()))
cost = list(map(int, input().split()))
adjacent = {x:[] for x in range(1, N + 1)}
for x in range(M):
  adjacent[nod1[x]].append((nod2[x], cost[x]))
for x in range(1, N + 1):
  ans = " ".join(f"({nod}, {wei})" for nod, wei in adjacent[x])
  print(f"{x}: {ans}")
```
