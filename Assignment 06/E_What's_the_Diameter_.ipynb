```python
from collections import deque
def BFS(edges, source, N):
  color = {x:0 for x in range(1, N + 1)}
  dist = [0] * N
  Q = deque()
  color[source] = 1
  Q.append(source)
  long_node = source
  while Q:
    vertex = Q.popleft()
    for x in edges[vertex]:
      if color[x] == 0:
        color[x] = 1
        dist[x - 1] = dist[vertex - 1] + 1
        Q.append(x)
        if dist[x - 1] > dist[long_node - 1]:
          long_node = x
  return long_node, dist[long_node - 1]

N = int(input())
edges = {x:[] for x in range(1, N + 1)}
for x in range(N - 1):
  node1, node2 = map(int, input().split())
  edges[node1].append(node2)
  edges[node2].append(node1)
X, diameter = BFS(edges, 1, N)
Y, diameter = BFS(edges, X, N)
print(diameter)
print(X, Y)
```
