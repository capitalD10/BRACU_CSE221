```python
import heapq
N = int(input())
def lexicoLatin(N):
  edges = {}
  indegree = {}
  dictionary = []
  for x in range(N):
    S = input()
    dictionary.append(S)
    for let in S:
      if let not in indegree:
        indegree[let] = 0
  for x in range(N - 1):
    word1 = dictionary[x]
    word2 = dictionary[x + 1]
    for ind in range(min(len(word1), len(word2))):
      if word1[ind] != word2[ind]:
        if word1[ind] not in edges:
          edges[word1[ind]] = []
        if word2[ind] not in edges[word1[ind]]:
          edges[word1[ind]].append(word2[ind])
          indegree[word2[ind]] += 1
        break
    else:
      if len(word1) > len(word2):
        return -1
  prioQ = []
  for x in indegree:
    if indegree[x] == 0:
      heapq.heappush(prioQ, x)
  valid_order = []
  while prioQ:
    vertex = heapq.heappop(prioQ)
    valid_order.append(vertex)
    if vertex in edges:
        for x in edges[vertex]:
          indegree[x] -= 1
          if indegree[x] == 0:
            heapq.heappush(prioQ, x)
  if len(valid_order) != len(indegree):
    return -1
  return "".join(valid_order)
print(lexicoLatin(N))
```
