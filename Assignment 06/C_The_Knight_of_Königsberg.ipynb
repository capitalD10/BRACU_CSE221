```python
from collections import deque
def BFSknight(N, x1, y1, x2, y2):
  Q = deque()
  Q.append((x1, y1, 0))
  color[x1 - 1][y1 - 1] = 1
  allowed = [(-2, -1), (-2, 1), (-1, -2), (-1, 2), (1, -2), (1, 2), (2, -1), (2, 1)]
  while Q:
    x, y, possible_move = Q.popleft()
    if x == x2 and y == y2:
      return possible_move
    for Mx, My in allowed:
      move1, move2 = x + Mx, y + My
      if (1 <= move1 <= N and 1 <= move2 <= N) and color[move1 - 1][move2 - 1] == 0:
        color[move1 - 1][move2 - 1] = 1
        Q.append((move1, move2, possible_move + 1))
  return -1

N = int(input())
x1, y1, x2, y2 = map(int, input().split())
color = [[0] * N for x in range(N)]
print(BFSknight(N, x1, y1, x2, y2))
```
