```python
from collections import deque
def Bipartite(edges, source, color):
  Q = deque()
  color[source] = 0
  roboHuman = [1, 0]
  Q.append(source)
  while Q:
    vertex = Q.popleft()
    for x in edges[vertex]:
      if color[x] == -1:
        color[x] = 1 - color[vertex]
        roboHuman[color[x]] += 1
        Q.append(x)
      elif color[x] == color[vertex]:
        return 0
  return max(roboHuman)

N, M = map(int, input().split())
edges = {x:[] for x in range(1, N + 1)}
color = {x:-1 for x in range(1, N + 1)}
players = 0
for x in range(1, M + 1):
  node1, node2 = map(int, input().split())
  edges[node1].append(node2)
  edges[node2].append(node1)
for x in range(1, N + 1):
  if color[x] == -1:
    players += Bipartite(edges, x, color)
print(players)
```
