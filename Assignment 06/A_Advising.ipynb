```python
from collections import deque
N, M = map(int, input().split())
edges = {x:[] for x in range(1, N + 1)}
indegree = [0] * (N)
for x in range(1, M + 1):
  node1, node2 = map(int, input().split())
  edges[node1].append(node2)
  indegree[node2 - 1] += 1
Q = deque()
for x in range(N):
  if indegree[x] == 0:
    Q.append(x + 1)
order = []
while Q:
  req = Q.popleft()
  order.append(req)
  for x in edges[req]:
    indegree[x - 1] -= 1
    if indegree[x - 1] == 0:
      Q.append(x)
if len(order) != N:
  print(-1)
else:
  print(*order)
```
