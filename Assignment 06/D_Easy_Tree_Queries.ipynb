```python
import sys
sys.setrecursionlimit(2 * 100000 + 5)
def subtree(N, R):
  edges = {x:[] for x in range(1, N + 1)}
  color = {x:0 for x in range(1, N + 1)}
  for x in range(N - 1):
    node1, node2 = map(int, input().split())
    edges[node1].append(node2)
    edges[node2].append(node1)
  subsize = {x:0 for x in range(1, N + 1)}
  def DFS(X, edges, color):
    color[X] = 1
    curr_size = 1
    for x in edges[X]:
      if color[x] == 0:
        curr_size += DFS(x, edges, color)
    subsize[X] = curr_size
    return curr_size
  subsize[R] = DFS(R, edges, color)
  queries = []
  for x in range(int(input())):
    queries.append(int(input()))
  for x in queries:
    print(subsize[x])

N, R = map(int, input().split())
subtree(N, R)
```
