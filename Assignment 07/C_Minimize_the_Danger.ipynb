```python
import heapq
N, M = map(int, input().split())
edges = {x:[] for x in range(1, N + 1)}
for x in range(M):
  node1, node2, weight = map(int, input().split())
  edges[node1].append((node2, weight))
  edges[node2].append((node1, weight))
danger_src = {x:float('inf') for x in range(1, N + 1)}
visited = {x:0 for x in range(1, N + 1)}
Q = [(0, 1)]
danger_src[1] = 0
while Q:
  cost, vertex = heapq.heappop(Q)
  visited[vertex] = 1
  for nod, wei in edges[vertex]:
    if visited[nod] == 0:
      if max(cost, wei) < danger_src[nod]:
        danger_src[nod] = max(cost, wei)
        heapq.heappush(Q, (danger_src[nod], nod))
for x in range(1, N + 1):
  if danger_src[x] == float('inf'):
    print(-1, end=" ")
  else:
    print(danger_src[x], end=" ")
```
