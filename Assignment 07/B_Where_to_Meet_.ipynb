```python
import heapq
def dijkstra(edges, source):
  dist_src = {x:float('inf') for x in range(1, N + 1)}
  visited = {x:0 for x in range(1, N + 1)}
  Q = [(0, source)]
  dist_src[source] = 0
  while Q:
    cost, vertex = heapq.heappop(Q)
    visited[vertex] = 1
    for nod, wei in edges[vertex]:
      if visited[nod] == 0:
        if dist_src[vertex] + wei < dist_src[nod]:
          dist_src[nod] = dist_src[vertex] + wei
          heapq.heappush(Q, (dist_src[nod], nod))
  return dist_src
N, M, S, T = map(int, input().split())
edges = {x:[] for x in range(1, N + 1)}
for x in range(M):
  node1, node2, weight = map(int, input().split())
  edges[node1].append((node2, weight))
if S == T:
  print(0, S)
else:
  Alice_dist = dijkstra(edges, S)
  Bob_dist = dijkstra(edges, T)
  minTime = dist = float('inf')
  for x in range(1, N + 1):
    if Alice_dist[x] != float('inf') and Bob_dist[x] != float('inf'):
      meet_time = max(Alice_dist[x], Bob_dist[x])
      if meet_time < minTime:
        minTime = meet_time
        dist = x
      elif meet_time == minTime and x < dist:
        dist = x
  if dist == float('inf'):
    print(-1)
  else:
    print(minTime, dist)
```
