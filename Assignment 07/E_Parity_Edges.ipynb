```python
import heapq
def Dijkstra(edges):
  dist_src = {x:[float('inf'), float('inf')] for x in range(1, N + 1)}
  Q = []
  for node, weight, parity in edges[1]:
    dist_src[node][parity] = weight
    heapq.heappush(Q, (weight, node, parity))
  while Q:
    wei, vertex, last_parity = heapq.heappop(Q)
    if dist_src[vertex][last_parity] < wei:
      continue
    for node, weight, current_parity in edges[vertex]:
      if current_parity == last_parity:
        continue
      if wei + weight < dist_src[node][current_parity]:
        dist_src[node][current_parity] = wei + weight
        heapq.heappush(Q, (dist_src[node][current_parity], node, current_parity))
  result = min(dist_src[N][0], dist_src[N][1])
  if result != float('inf'):
    return result
  return -1

N, M = map(int, input().split())
node1 = list(map(int, input().split()))
node2 = list(map(int, input().split()))
weight = list(map(int, input().split()))
edges = {x:[] for x in range(1, N + 1)}
for x in range(M):
  edges[node1[x]].append((node2[x], weight[x], weight[x] % 2))
print(Dijkstra(edges))
```
