```python
import heapq
def SecondDijkstra(N, M, S, D, edges):
  best_dist = [float('inf')] * (N + 1)
  second_dist = [float('inf')] * (N + 1)
  Q = [(0, S)]
  best_dist[S] = 0
  while Q:
    cost, vertex = heapq.heappop(Q)
    for node, wei in edges[vertex]:
      if cost + wei < best_dist[node]:
        second_dist[node] = best_dist[node]
        best_dist[node] = cost + wei
        heapq.heappush(Q, (best_dist[node], node))
      elif best_dist[node] < (cost + wei) < second_dist[node]:
        second_dist[node] = cost + wei
        heapq.heappush(Q, (second_dist[node], node))
  if second_dist[D] != float('inf'):
    return second_dist[D]
  return -1

N, M, S, D = map(int, input().split())
edges = [[] for x in range(N + 1)]
for x in range(M):
  node1, node2, weight = map(int, input().split())
  edges[node1].append((node2, weight))
  edges[node2].append((node1, weight))
print(SecondDijkstra(N, M, S, D, edges))
```
