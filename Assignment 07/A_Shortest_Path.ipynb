```python
import heapq
N, M, S, D = map(int, input().split())
node1 = list(map(int, input().split()))
node2 = list(map(int, input().split()))
weight = list(map(int, input().split()))
edges = {x:[] for x in range(1, N + 1)}
for x in range(M):
  edges[node1[x]].append((node2[x], weight[x]))
dist_src = {x:float('inf') for x in range(1, N + 1)}
visited = {x:0 for x in range(1, N + 1)}
parent = {x:0 for x in range(1, N + 1)}

Q = [(0, S)]
dist_src[S] = 0
while Q:
  cost, vertex = heapq.heappop(Q)
  visited[vertex] = 1
  for nod, wei in edges[vertex]:
    if visited[nod] == 0:
      if dist_src[vertex] + wei < dist_src[nod]:
        dist_src[nod] = dist_src[vertex] + wei
        parent[nod] = vertex
        heapq.heappush(Q, (dist_src[nod], nod))
if dist_src[D] == float('inf'):
  print(-1)
else:
  result = []
  last = D
  while last != 0:
    result.append(last)
    last = parent[last]
  print(dist_src[D])
  print(*result[::-1])
```
