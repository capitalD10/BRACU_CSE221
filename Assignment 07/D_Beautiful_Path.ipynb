```python
import heapq
def Dijkstra(edges, S, weights, D):
  dist_src = {x:float('inf') for x in range(1, N + 1)}
  Q = [(weights[S - 1], S)]
  dist_src[S] = weights[S - 1]
  while Q:
    wei, vertex = heapq.heappop(Q)
    if dist_src[vertex] < wei:
      continue
    if vertex == D:
      return wei
    for x in edges[vertex]:
      if dist_src[vertex] + weights[x - 1] < dist_src[x]:
        dist_src[x] = dist_src[vertex] + weights[x - 1]
        heapq.heappush(Q, (dist_src[x], x))
  return -1
N, M, S, D = map(int, input().split())
weights = list(map(int, input().split()))
edges = {x:[] for x in range(1, N + 1)}
for x in range(M):
  node1, node2 = map(int, input().split())
  edges[node1].append(node2)
print(Dijkstra(edges, S, weights, D))
```
