```python
def searcher(n, arr, num):
  Left = 0
  Right = n - 1
  index = n
  while Left <= Right:
    mid = (Left + Right) // 2
    if arr[mid] >= num:
      index = mid
      Right = mid - 1
    else:
      Left = mid + 1
  return index

n, q = list(map(int, input().split()))
arr = list(map(int, input().split()))
rng = [list(map(int, input().split())) for x in range(q)]
for x in rng:
  print(searcher(n, arr, x[1] + 1) - searcher(n, arr, x[0]))
```
