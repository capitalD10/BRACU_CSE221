```python
N = int(input())
arr1 = list(map(int, input().split()))
M = int(input())
arr2 = list(map(int, input().split()))
i = j = 0
res = []
while i < N and j < M:
    if arr1[i] < arr2[j]:
        res.append(arr1[i])
        i += 1
    else:
        res.append(arr2[j])
        j += 1

if i < N:
  res.extend(arr1[i:])
elif j < M:
  res.extend(arr2[j:])
print(" ".join(map(str, res)))
```
