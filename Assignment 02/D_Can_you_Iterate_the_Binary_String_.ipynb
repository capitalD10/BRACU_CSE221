```python
def binaryOne(S):
  l = 0
  r = len(S) - 1
  where = -1
  while l <= r:
      mid = (l + r) // 2
      if S[mid] == '1':
          where = mid + 1
          r = mid - 1
      else:
          l = mid + 1
  return where
T = int(input())
res = []
for x in range(T):
  S = input()
  res.append(binaryOne(S))
for x in res:
  print(x)
```
