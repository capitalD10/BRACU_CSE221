```python
def twoSum(N, arr):
    first = 0
    last = N[0] - 1
    while first < last:
        sum = arr[first] + arr[last]
        if sum < N[1]:
            first += 1
        elif sum > N[1]:
            last -= 1
        else:
            return first + 1, last + 1
    return [-1]

N = list(map(int, input().split()))
arr = list(map(int, input().split()))
print(*twoSum(N, arr))
```
